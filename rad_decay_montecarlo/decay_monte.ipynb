```python
import numpy as np
import matplotlib.pyplot as plt
```

**Simulation Parameters**

- Total number of atoms: Initial number of radioactive atoms N0.
- Decay probability: Probability of an atom decaying in each time step.
- Time step: Duration of each step in seconds.
- Total time: Total simulation duration in seconds.

```python
num_atoms = 10000  # number of atoms
decay_probability = 0.1  # probability of decay per time step
time_step = 1  # time step duration (s)
total_time = 100  # total simulation time (s)
```

**Monte Carlo Simulation**

1. At each time step, every atom has a probability of decaying.
2. The number of remaining atoms is updated and recorded.
3. The simulation runs for the total duration, divided into discrete time steps.

```python
# Initialize variables
remaining_atoms = num_atoms
remaining_counts = []
time_steps = int(total_time / time_step)

# Monte Carlo simulation
for t in range(time_steps):
    decayed = 0
    for _ in range(remaining_atoms):
        if np.random.random() < decay_probability:
            decayed += 1
    remaining_atoms -= decayed
    remaining_counts.append(remaining_atoms)
```

**Results and Visualization**

- The red dots represent the number of remaining atoms at each time step based on the Monte Carlo simulation.
- The dashed blue line shows the expected exponential decay curve for comparison.


```python
# Generate time values for the x-axis
time_values = np.linspace(0, total_time, len(remaining_counts)) 

# Calculate the expected decay curve
expected_curve = num_atoms * np.exp(-decay_probability * time_values)

# Plot the results
plt.figure(figsize = (10, 6))

# Scatter plot for simulation
plt.scatter(time_values,
            remaining_counts,
            color='red',
            s = 40,
            label='Simulated Remaining Atoms')

# Plot the expected decay curve
plt.plot(time_values,
         expected_curve,
         color = 'blue',
         linestyle = '--',
         linewidth = 2,
         label = 'Expected Decay Curve')


plt.title('Monte Carlo Simulation of Radioactive Decay')
plt.xlabel('Time (s)')
plt.ylabel('Number of Remaining Atoms')

plt.legend()
plt.grid()
plt.show()
```
