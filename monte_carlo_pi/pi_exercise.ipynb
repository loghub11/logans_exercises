```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_points = 10000 
```

**Setting Parameters**

This cell sets num_points to the total number of random points to be generated. A higher number of points usually means a more accurate estimate for π.

```python
# Monte Carlo Method to Estimate Pi
inside_circle = 0 
points_x = []
points_y = []
```

**Initializing Variables**

* inside_circle will count the points that fall inside the circle.
* points_x and points_y are lists to store the x and y coordinates of each point. 

```python
for _ in range(num_points):
    x, y = np.random.uniform(-1, 1), np.random.uniform(-1, 1)
    points_x.append(x)
    points_y.append(y)
    if x**2 + y**2 <= 1:
        inside_circle += 1
```

**Generating Random Points and Counting Points in the Circle**

For each point:
1. generate a random x and y coordinate between -1 and 1.
2. check if the point falls within the unit circle by calculating if (x^2 + y^2 <= 1), if it does, increase the inside_circle counter.
3. add each point’s coordinates to points_x and points_y so we can plot them later.

```python
pi_estimate = (inside_circle / num_points) * 4
```

**Calculating Pi Estimate**

Now to calculate an estimate for π. The basic idea is to use the ratio of points inside the circle to the total points in the square.


```python
plt.figure(figsize = (8, 8))
plt.scatter(points_x, points_y, color = ['blue' if x**2 + y**2 <= 1 else 'red' for x, y in zip(points_x, points_y)], s = 1)
plt.title(f"Monte Carlo Estimation of Pi \n Estimated Pi = {pi_estimate:.6f}")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.grid()
plt.show()
```

**Visualization of Points**

* **Blue points** represent those that fall within the unit circle.
* **Red points** are outside the circle.

The plot title displays the calculated value for π, which gives a quick look at how well this method approximates π.
